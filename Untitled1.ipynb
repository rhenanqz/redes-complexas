```python
from numpy import *
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.Graph()
G.add_node('1')
G.add_node('2')
G.add_node('3')
G.add_node('4')
G.add_node('5')
G.add_node('6')

G.add_edge('1','2')
G.add_edge('1','5')
G.add_edge('2','3')
G.add_edge('2','5')
G.add_edge('3','4')
G.add_edge('4','5')
G.add_edge('4','6')

pos=nx.spring_layout(G)
nx.draw(G, with_labels = True, node_size=500, font_size=16, pos = pos)
plt.show(True)
```
